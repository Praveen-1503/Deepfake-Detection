```python
import os
from PIL import Image

# ‚úÖ Base directory for your testing images
BASE_DIR = "/Users/praveen1517/Documents/LiveInLab - III/RealReveal/backend/testing images"

# ‚úÖ Subfolders to process
folders = ["fake", "real"]

# ‚úÖ Target image size for your model
TARGET_SIZE = (224, 224)

def resize_images_in_folder(folder_path):
    for root, _, files in os.walk(folder_path):
        for file in files:
            if file.lower().endswith((".jpg", ".jpeg", ".png")):
                img_path = os.path.join(root, file)
                try:
                    with Image.open(img_path) as img:
                        img = img.convert("RGB")
                        img = img.resize(TARGET_SIZE)
                        img.save(img_path, quality=95)
                    print(f"‚úÖ Resized: {img_path}")
                except Exception as e:
                    print(f"‚ö†Ô∏è Error resizing {img_path}: {e}")

# ‚úÖ Process both fake and real folders
for folder in folders:
    folder_path = os.path.join(BASE_DIR, folder)
    resize_images_in_folder(folder_path)

print("\nüéâ All images in 'testing images' resized to 224x224 successfully!")
```
